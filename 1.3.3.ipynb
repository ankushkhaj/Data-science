```python
import pandas as pd
players = pd.read_csv("C:/Personal/09142640/Desktop/ATP_Rankings_updated.csv")
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
players.head()
```

```python
winners = players.loc[(players['Round']=='F') & (players['surface']=='Hard')]
winner_name1=pd.DataFrame(winners)
plt.xticks(rotation='vertical')
sns.countplot(winner_name1['winner_name'])
plt.ylabel("Num of Tournaments won")
plt.xlabel("Player name")
#plt.hist(winners, alpha=.5,label='winners') 
#plt.xlabel('Winners')
#plt.legend(loc='upper right')
plt.title('Plot 1: Winners won in Final ')
#winner_name1
```

```python
Pros:This plot shows the winner names and how many tournaments they have won

Cons:Other information such as year, country etc are missing in this plot
```

```python
plt.title('Plot 2: Winners rank ')
sns.distplot(winners['winner_rank'])
```

Pros:This histogram represents distribution of data (i.e rank here) by forming bins along the range of the data 
and then drawing bars to show the number of observations that fall in each bin.

Cons:It accepts only 1 variable ,here it is only rank

```python
sns.jointplot(x='year', y='winner_rank', data=winners)
```

Pros:This plot shows the rank of the players and their respective year

Cons:This plot is a little misleading and information explained in the plot is not clear where it has scatterplot and histogram

```python
plt.title('Plot 2: Winners rank ')
sns.boxplot(x='year', y='winner_rank', data=winners)
```

Pros:This plot explains the rank of the winners and their respective year
Cons:This plot is a little confusing  if a person looks at it first time since we should have basic knowledge of how to read box plot

```python
g = sns.lmplot(x='winner_ht', # Variable 1.
               y='winner_rank', # Variable 2.
               data=winners, # Data
               fit_reg=False, # If set to true, plots a regression line.
               scatter_kws={'alpha':0.6}) # Set points to semi-transparent to see overlaping points.
g.set_ylabels("winner rank")
g.set_xlabels("winner height")
plt.title('Scatterplot: Winner height and  their ranks ')
plt.show()

```

Pros:This plot explains the rank of the winners and the height of the players

Cons:This plot doesnt provide us enough information to make a conclusion

```python
g = sns.lmplot(x='winner_ht', # Variable 1.
               y='winner_rank', # Variable 2.
               data=winners, # Data
               fit_reg=True, # If set to true, plots a regression line.
               scatter_kws={'alpha':0.6}) # Set points to semi-transparent to see overlaping points.
g.set_ylabels("winner rank")
g.set_xlabels("winner height")
plt.title('Scatterplot: Winner height and  their ranks ')
plt.show()
```

```python
Pros:This plot explains the rank of the winners and the height of the players

Cons:This plot doesnt provide us enough information to make a conclusion
```

```python
plt.bar( winners['winner_ht'],winners['winner_rank'],
                 alpha=0.5,
                 color='b'
                 )
plt.ylabel('Winner rank')
plt.xlabel('Winner height')
plt.title('Winner rank along with their height')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
#plt.legend(loc='upper right')
```

Pros:This plot explains the rank of the winners and the height of the players

Cons:This plot doesnt provide us enough information to make a conclusion

```python
sns.boxplot(winners['winner_ht'],winners['winner_rank'])
plt.title('Plot 3: Winner rank along with their height')
sns.despine(offset=10, trim=True)
plt.show()
```

Pros:This plot explains the rank of the winners and the height of the players

Cons:This plot doesnt provide us enough information to make a conclusion.It is hard to read as well

```python
g = sns.factorplot('year','winner_rank', hue="winner_ioc", data=winners,
                   size=6, kind="point", palette="pastel",ci=95,dodge=True,join=False)
g.despine(left=True)
g.set_ylabels("Year")
g.set_xlabels("Rank")
plt.title('Pointplot: rank with year')
plt.show()
```

















Pros:This plot explains the rank of the winners with the years of the players and their respective country 

Cons:This plot doesnt provide us enough information to make a conclusion.It is hard to read as well


```python
sns.set(style="white")

g = sns.lmplot(x='winner_rank',y='winner_age',
               data=winners, # Data.
               hue='winner_ioc',col='year',
               fit_reg=True,
               ci=False,scatter_kws={'alpha':0.4})               
g.set_ylabels("Age")
g.set_xlabels("Rank")
plt.suptitle('Winner rank with their age ')
plt.subplots_adjust(top=0.9)
plt.show()
```

Pros:This plot explains the rank of the winners with the agw of the players and their respective country 

Cons:This plot doesnt provide us enough information to make a conclusion.It is hard to read as well
