Using Logistic Regression model below where the output variable is Murder_cat and the input variables are rest of all the columns.We will use test and training dataset to validate and fit the data 

```python
lr = LogisticRegression(C=1e9)
y = data['Murder_Cat']
X = data.drop(columns='Murder_Cat')
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
# Peprocessing X's to center to the mean
X = scale(X)

# Creating train and test variables
# Creating decision tree using the training data
#decision_tree = tree.DecisionTreeRegressor()
lr.fit(X_train, y_train)
print('Training set: ' + str(lr.score(X_train, y_train)))
print('Test set: ' + str(lr.score(X_test, y_test)))

```
