# About the data
The data presented here is a collection of all tennis players played in ATP Tournaments.This data explains all the information such as Tournament played, player name, rank of the player,Surface where the tournaments are being played,tournament year, nationality of the player and some other useful information.The data presented  provides enough information to perform  analysis to get a valuable insights.The data can be downloaded from kaggle here-https://www.kaggle.com/sijovm/atpdata



```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
df=pd.read_csv("C:/Personal/09142640/Desktop/ATP_Rankings_updated.csv")
df.head()
plt.subplot(1, 2, 1)
counts=df['winner_name'].value_counts()
df_count = pd.DataFrame(counts)
df_counts=counts.reset_index()
Top_5=df_counts.sort_values('winner_name',ascending=False).head(5)
plt.plot(Top_5['winner_name'],Top_5['index'],linestyle='', marker='o')
plt.title('Top 5 Tennis Players based on Total Matches won since 2015')
plt.xlabel('Matches won',fontsize=18)
plt.ylabel('Player',fontsize=16)
plt.show()
```

# Q1:Which players have won the most matches since 2015?

```python
#Q:What has been the peak year for top5 playters based on the highest tournaments won since 2015
#A:I have created sub-plot to show that some players had their peak year of 2016 and some of them had peak year of 2017
counts=df_Final['winner_name'].value_counts()
#counts
#df_count = pd.DataFrame(counts)
#count=df_count[columns='count']
df_counts=counts.to_frame().reset_index()
df_counts.columns = ['winner_name', 'Championship won']
Top_5=df_counts.sort_values('Championship won',ascending=False).head(5)
Top_5
df3= df.loc[df['winner_name'] .isin(Top_5['winner_name'])]
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
df3[['winner_name','year']]
plt.plot(df3['year'],df3['winner_name'],linestyle=' ', marker='o')
plt.title('Top 5 Tennis Players with the year of Championship wins')
plt.xlabel('Year',fontsize=18)
plt.ylabel('Player',fontsize=16)
plt.show()
plt.subplot(1, 2, 2)
df3[['winner_name','winner_age']]
plt.plot(df3['winner_age'],df3['winner_name'],linestyle=' ', marker='o')
plt.title('Top 5 Tennis Players with their age during their championship wins')
plt.xlabel('Age',fontsize=18)
plt.ylabel('Player',fontsize=16)
plt.show()
```

# Q5:Which player has been consistent among all tennis surfaces and which surface has been the favourite and least favourite?
Ans:The above chart shows that Dominic Thiem has won his champhionships in all the surfaces-Hard,Clay and Grass,so he has the been the most consistent in the all the tennis surfaces.Most of  players have won on Clay and Hard and very few on Grass

```python
plt.title('Top 5 Tennis Players with their mean age of winning championships and their Standard Deviation')
plt.hist(df_surface['winner_age'], bins=20, color='c')

# Add a vertical line at the mean.
plt.axvline(df_surface['winner_age'].mean(), color='r', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(df_surface['winner_age'].mean() + df_surface['winner_age'].std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(df_surface['winner_age'].mean()-df_surface['winner_age'].std(), color='b', linestyle='dashed', linewidth=2) 
plt.ylabel('Number of Championship won',fontsize=14)
plt.xlabel('Player Age',fontsize=14)
# Print the histogram.
plt.show()
```
