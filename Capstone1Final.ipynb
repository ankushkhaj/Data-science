# About the data
The data presented here is a collection of all tennis players played in ATP Tournaments.This data explains all the information such as Tournament played, player name, rank of the player,Surface where the tournaments are being played,tournament year, nationality of the player and some other useful information.The data presented  provides enough information to perform  analysis to get a valuable insights.The data can be downloaded from kaggle here-https://www.kaggle.com/sijovm/atpdata



```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#set low_memory=False
%matplotlib inline
plt.figure(figsize=(40,10))
df=pd.read_csv("C:/Personal/09142640/Desktop/ATP_Rankings_updated.csv")
df.head()
plt.subplot(1, 2, 1)
#counts=df.groupby('Player')['Player'].count()
counts=df['winner_name'].value_counts()
#counts
df_count = pd.DataFrame(counts)
#count=df_count[columns='count']
df_counts=counts.reset_index()
df_counts
Top_5=df_counts.sort_values('winner_name',ascending=False).head(5)
Top_5

plt.plot(Top_5['winner_name'],Top_5['index'],linestyle='', marker='o')
plt.title('Top 5 Tennis Players based on Total Matches won since 2015')
plt.xlabel('Matches won')
plt.ylabel('Player')
plt.show()
```

The  subplot above  shows  that Novak Djokovic is leading for the most matches won since 2015 followed by Domonic Thiem and so on.Here y axis represents the player name and x axis represents the number of matches won by each player since 2015.The plot above is a clear indicator of which player has been peforming better than the remaining players

```python
#Q:What has been the peak year for top5 playters based on the highest tournaments won since 2015
#A:I have created sub-plot to show that some players had their peak year of 2016 and some of them had peak year of 2017
counts=df_Final['winner_name'].value_counts()
#counts
#df_count = pd.DataFrame(counts)
#count=df_count[columns='count']
df_counts=counts.to_frame().reset_index()
df_counts.columns = ['winner_name', 'Championship won']
Top_5=df_counts.sort_values('Championship won',ascending=False).head(5)
Top_5
df3= df.loc[df['winner_name'] .isin(Top_5['winner_name'])]
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
df3[['winner_name','year']]
plt.plot(df3['year'],df3['winner_name'],linestyle=' ', marker='o')
plt.title('Top 5 Tennis Players with the year of Championship wins')
plt.xlabel('Year')
plt.ylabel('Player')
plt.show()
plt.subplot(1, 2, 2)
df3[['winner_name','winner_age']]
plt.plot(df3['winner_age'],df3['winner_name'],linestyle=' ', marker='o')
plt.title('Top 5 Tennis Players with their age during their championship wins')
plt.xlabel('Age')
plt.ylabel('Player')
plt.show()
```

```python

#Q:Which player has been best performing on which surface based on maximum number tournaments won for top5 players
#A:I have created subplot to show that each player has their respective favourite surface to play at
df_surface=df3[df.Round=='F']
df_surface[['winner_name','surface']]
plt.plot(df_surface['surface'],df_surface['winner_name'],linestyle=' ', marker='o')
plt.title('Top 5 Tennis Players with their fav surface where they have won championships')
plt.xlabel('Surface')
plt.ylabel('Player')
plt.show()
```

```python
#For age related data for tennis players, values are not normally disrtibuted.the mean is still placed where the data clusters, but the cluster is not centered, and 
#the standard deviation does not encompass the same amount of data on each side of the mean. 
#Put another way, for age related data for tennis players, the mean is no longer a measure of "central" tendency, 
#as it does not fall in the center, and the standard deviation no longer describes how much variance there is.
#Plot a histogram for age of .
plt.title('Top 5 Tennis Players with their mean age of winning championships and their Standard Deviation')
plt.hist(df_surface['winner_age'], bins=20, color='c')

# Add a vertical line at the mean.
plt.axvline(df_surface['winner_age'].mean(), color='r', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(df_surface['winner_age'].mean() + df_surface['winner_age'].std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(df_surface['winner_age'].mean()-df_surface['winner_age'].std(), color='b', linestyle='dashed', linewidth=2) 
plt.ylabel('Number of Championship won')
plt.xlabel('Player Age')
# Print the histogram.
plt.show()
```

The above plot shows  the mean age of winning championships for top5 players with one standard deviation above the mean and one standard deviation below the mean.The mean has been shown with solid line and the dotted lines are for the standard deviations.
