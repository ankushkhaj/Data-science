```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#set low_memory=False
%matplotlib inline
plt.figure(figsize=(40,10))
df=pd.read_csv("C:/Personal/09142640/Desktop/ATP_Rankings_updated.csv")
df
```

```python
#For age related data for tennis players, values are not normally disrtibuted.the mean is still placed where the data clusters, but the cluster is not centered, and 
#the standard deviation does not encompass the same amount of data on each side of the mean. 
#Put another way, for age related data for tennis players, the mean is no longer a measure of "central" tendency, 
#as it does not fall in the center, and the standard deviation no longer describes how much variance there is.
#Plot a histogram for age of .
plt.hist(df_surface['winner_age'], bins=20, color='c')

# Add a vertical line at the mean.
plt.axvline(df_surface['winner_age'].mean(), color='r', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(df_surface['winner_age'].mean() + df_surface['winner_age'].std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(df_surface['winner_age'].mean()-df_surface['winner_age'].std(), color='b', linestyle='dashed', linewidth=2) 

# Print the histogram.
plt.show()
```

```python
#Q: Check the variability of 2 random age samples from the data set
#A:we can see that few of the values in the sample groups overlap.
#The groups are so variable that the distribution of values is quite broad
plt.hist(sample1,  bins=15,alpha=0.5,label='sample 1') 
plt.hist(sample2,  bins=15,alpha=0.5,label='sample 2') 
plt.legend(loc='upper right') 
plt.show()
```

```python
#Q: How to check if the 2 sample data sets have no statistically significant difference between the sample 
#data  sets
#A:We can use t-test to vefity the above theory.The t test (also called Student’s T Test) compares two averages (means) 
#and tells you if they are different from each other. It lets you know if those differences could have happened by chance.
#Here  p value of 0.19 which is 19% which is very high and more than 5%,so we are not going to reject null hypothesis.
#Hence,there is no  statistically significant difference between these 2 sample sets
from scipy.stats import ttest_ind

print(ttest_ind(sample2, sample1, equal_var=False))
```
